```python
import numpy as np

np.random.seed(42)
lons = np.random.uniform(-180, 180, 100)
lats = np.random.uniform(-90, 90, 100)
temperature = np.random.normal(20, 10, 100)
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

fig = plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.Orthographic(central_longitude=np.mean(lons), central_latitude=np.mean(lats)))
ax.coastlines()
ax.stock_img()

# Plot synthetic data on the map
scatter = ax.scatter(lons, lats, transform=ccrs.Geodetic(), c=temperature, cmap='coolwarm',alpha=0.5)
plt.colorbar(scatter)
plt.title('Synthetic Temperature Data')
plt.show()
```
