```python
import numpy as np
from scipy.interpolate import splprep, splev

# Define the x and y coordinates of the river path
x = np.linspace(0,10,100)
y = np.sin(x/2)

# Add noise to the x and y coordinates to create a more natural-looking path
x_noise = x + np.random.normal(0, 0.2, len(x))
y_noise = y + np.random.normal(0, 0.2, len(y))

# Use splprep to create a smooth spline representation of the path
tck, u = splprep([x_noise, y_noise], s=0.05)

# Evaluate the spline at a set of evenly spaced points
u_new = np.linspace(u.min(), u.max(), 1000)
x_new, y_new = splev(u_new, tck)

# Create a 3D representation of the river by adding a z coordinate
z = np.zeros_like(x_new)

# Create a mesh representation of the river
river_mesh = np.column_stack((x_new, y_new, z))

fig = go.Figure(data=[go.Scatter3d(x=x_new, y=y_new, z=z, mode='lines', line=dict(width=2, color='lightblue'))])
fig.update_layout(scene = dict(
                    xaxis_title='X',
                    yaxis_title='Y',
                    zaxis_title='Z'))
fig.show()

```

```python
import noise
import numpy as np
import matplotlib.pyplot as plt

# Define the size of the topography
nx, ny = 100, 100

# Create a 2D array of fractal noise values
elevation = np.zeros((nx, ny))
for i in range(nx):
    for j in range(ny):
        elevation[i, j] = noise.pnoise2(i/10, j/10, octaves=5, persistence=0.5, lacunarity=2)

# Use the `imshow` function from `matplotlib` to display the topography
plt.imshow(elevation, cmap='terrain')
plt.colorbar()
plt.show()
```
